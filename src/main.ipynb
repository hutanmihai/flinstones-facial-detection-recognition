## 1. Importing Libraries

```python
import numpy as np
import cv2 as cv
```

```python
class Annotation:
    def __init__(self, path, xmin, ymin, xmax, ymax, character):
        self.character = character
        self.xmin = xmin
        self.ymin = ymin
        self.xmax = xmax
        self.ymax = ymax


class Image:
    def __init__(self, path, name, annotations: list[Annotation]):
        self.path = path
        self.name = name
        self.annotations = annotations
        self.image = cv.imread(path)
```
